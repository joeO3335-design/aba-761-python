```python
# Create a 1D numpy array that contains all even numbers up to 10,000
even_numbers = np.arange(2, 10001, 2)
# OR
even_numbers = np.linspace(2, 10000, 5000, dtype=int)
# OR
even_numbers = np.array([x for x in range(2, 10001, 2)])
even_numbers
```

```python
# Create a 2D numpy array where one array contains all evens up to 1,000
# and the second array contains the square of the first array.

# Create the first array (even numbers up to 1,000)
evens = np.arange(2, 1001, 2)

# Create the second array (square of the even numbers)
squares = evens ** 2

# Stack them into a 2D array (2 rows, N columns)
result = np.vstack([evens, squares])
# OR
result = np.column_stack((evens, squares))  # Creates a (500, 2) array

result
```

### Simple functions

```python
# Write a simple function that adds two numbers together with proper
# documentation. Demonstrate its use.
def add_numbers(num1, num2):
  """
  Add two numbers together

  Returns
  ------
  the_sum : type of arguments
  """
  return num1 + num2

add_numbers(77, 1224)
```

```python
# Write a simple function that multiplies two numbers
# together with proper documentation. Demonstrate its use.
def multiply_numbers(num1, num2):
  """
  Multiplies two numbers together

  Returns
  ------
  the_product : type of arguments
  """
  return num1*num2

multiply_numbers(77, 1224)
```

```python
# Write a simple function called 'integer_division' that
# divides a number but returns only the
# whole integer value, not any fractional remainder.
# Demonstrate its use with 10 divided by 3, and 25 divided by 4.
def integer_division(a, b):
    """
    Divides two numbers but returns only the whole integer value.

    Returns
    ------
    quotient : type of arguments
    """
    return a // b  # Floor division operator discards the remainder

result = integer_division(10, 3)
print(result)  # Expected output: 3

result = integer_division(25, 4)
print(result)  # Expected output: 6
```

```python
# Raise the output of integer_division for 25 divided by 4 to the third power
# and store it as a new object called "new_result".
# Show in code how you'd test whether new_result equals 260.
new_result = result**3
new_result == 260
```

### Advanced: Tying some things together

```python
# Create a dataframe with 10 rows and three columns.
# The first column should be a sequence of dates between today and ten days from today.
# The second column should be the number of trials run during an ABA session.
#   - Use a random draw from a Gaussian distribution averageing 60
#     with a standard deviation of 10
# The third column should be the percentage of trials with correct responding.
#   - Use a random draw from uniform distribution with a low of 0 and high of 100.

# Generate a sequence of 10 dates starting from today
date_range = pd.date_range(start=pd.Timestamp.today().normalize(), periods=10, freq='D')

# Generate the number of trials using a Gaussian distribution (mean=60, std=10)
num_trials = np.random.normal(loc=60, scale=10, size=10).astype(int)  # Convert to integers

# Generate the percentage of correct responses using a uniform distribution (0 to 100%)
percent_correct = np.random.uniform(low=0, high=100, size=10)

# Create the DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Num_Trials': num_trials,
    'Percent_Correct': percent_correct
})
df
```
