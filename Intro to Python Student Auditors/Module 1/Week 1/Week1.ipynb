```python
import streamlit as st # type: ignore

st.title("Revenue Calculator App")

st.markdown("### Inputs")

rate_per_day = st.number_input(
    "Rate Per Day (per student)",
    min_value=0.0,
    value=350.0,
    step=10.0
)

number_of_students = st.number_input(
    "Number of Students",
    min_value=0,
    value=25,
    step=1
)

days_per_week = st.number_input(
    "Days Per Week",
    min_value=1,
    max_value=7,
    value=5
)

weeks_per_year = st.number_input(
    "Weeks Per Year",
    min_value=1,
    value=52
)

weeks_per_month = st.number_input(
    "Weeks Per Month (Multiplier)",
    min_value=1.0,
    value=4.5,
    step=0.1
)

st.markdown("---")
st.markdown("### Revenue Output")

daily_gross = rate_per_day * number_of_students
weekly_gross = daily_gross * days_per_week
monthly_gross = weekly_gross * weeks_per_month
yearly_gross = weekly_gross * weeks_per_year

st.metric("Daily Gross", f"${daily_gross:,.2f}")
st.metric("Weekly Gross", f"${weekly_gross:,.2f}")
st.metric("Monthly Gross", f"${monthly_gross:,.2f}")
st.metric("Yearly Gross", f"${yearly_gross:,.2f}")
```
