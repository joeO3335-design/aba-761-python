### NumPy Basics

```python
# Create a 1D NumPy array of the first 10 natural numbers.
import numpy as np
array_1d = np.arange(1, 11)
print(array_1d)
```

```python
# Print the 3rd and 7th element of the array you just created
print(array_1d[2])
print(array_1d[6])
```

```python
# Create a 1D numpy array that contains all even numbers up to 10,000
even_numbers = np.arange(2, 10001, 2)
print(even_numbers)
```

```python
# Create a 2D numpy array where one array contains all evens up to 1,000
# and the second array contains the square of the first array.
evens = np.arange(2, 1001, 2)
squares = evens ** 2
array_2d = np.array([evens, squares])
print(array_2d)
```

### Pandas DataFrame

```python
# Create a pandas DataFrame with two columns: 'Name' and 'Age',
# with at least three entries, and using dictionaries to populate the columns.
import pandas as pd
data = {
    "Name": ["Fred", "Montrose", "Ogden"],
    "Age": [25, 44, 32]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Create a pandas DataFrame with two columns: 'Name' and 'Age',
# with at least three entries, and using lists to populate the columns.
import pandas as pd
name = ["Harry", "Ron", "Severus"]
age = [14, 13, 15]
df = pd.DataFrame({"Name" : name, "Age" :age})
print(df)
```
