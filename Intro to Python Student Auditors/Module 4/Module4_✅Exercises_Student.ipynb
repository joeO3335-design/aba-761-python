```python
# Define a custom function and apply it to a column.(📙Page )
df = pd.DataFrame({'scores': [85, 92, 78, 60, 95]})
def grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'F'
df['grade'] = df['scores'].apply(grade)
graded_df = df
print(graded_df)
```

```python
# Create a new categorical column and group the DataFrame by this column.(📙Page )
import pandas as pd

df = pd.DataFrame({
    'A': [10, 20, 30],
    'B': [40, 50, 60],
    'C': [70, 80, 90]
})

def categorize_B(x):
    if x >= 55:
        return 'High'
    elif x >= 45:
        return 'Medium'
    else:
        return 'Low'

gimme_shelter = df.copy()
gimme_shelter["B_category"] = gimme_shelter["B"].apply(categorize_B)

print(gimme_shelter)
```

```python
# Compute the sum of column 'A' for each group.(📙Page )
gimme_shelter
print(gimme_shelter.groupby("B_category")["A"].mean())
```

```python
# Reshape a DataFrame from long to wide format using pivot.(📙Page 277)
import pandas as pd

df_long = pd.DataFrame({
    'name':     ['Alice', 'Alice', 'Bob', 'Bob'],
    'subject':  ['Math', 'Science', 'Math', 'Science'],
    'score':    [85, 92, 78, 88]
})

df_wide = df_long.pivot(index='name', columns='subject', values='score').reset_index()
df_wide.columns.name = None

print(df_long)
print()
print(df_wide)
```

```python
# Melt a wide DataFrame back into long format.(📙Page 277)
import pandas as pd

df_long = pd.DataFrame({
    'name':     ['Alice', 'Alice', 'Bob', 'Bob'],
    'subject':  ['Math', 'Science', 'Math', 'Science'],
    'score':    [85, 92, 78, 88]
})

df_wide = df_long.pivot(index='name', columns='subject', values='score').reset_index()
df_wide.columns.name = None

df_melted = df_wide.melt(id_vars='name', var_name='subject', value_name='score')

print(df_wide)
print()
print(df_melted)
```
