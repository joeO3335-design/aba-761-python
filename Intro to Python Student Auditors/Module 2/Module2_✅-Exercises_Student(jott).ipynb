```python
# Write a function greet() that prints "Hello, welcome to Python!" when called. 
# Call the function.
def greet():
    print("Hello, welcome to Python!")
greet()
```

```python
# Write a function is_even(n) that takes a number and returns 
# True if it's even and False if it's odd.
def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False
print(is_even(4))
print(is_even(7))
```

```python
# Write a function celsius_to_fahrenheit(celsius) that takes a 
# temperature in Celsius and returns the Fahrenheit equivalent. 
# Use the formula: Fahrenheit = (Celsius x 9/5) + 32
def celcius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit
```

```python
# Write a function that checks if a given number exists in a tuple (5, 10, 15, 20, 25)
def check_number(n):
    numbers = (5, 10, 15, 20, 25)
    if n in numbers:
        return True
    else:
        return False
print(check_number(10))
print(check_number(7))
```

```python
# Write a function max_of_three(a, b, c) that takes three numbers and returns the largest one.
def max_of_three(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b>= c:
        return b
    else: 
        return c
print(max_of_three(33, 35, 52))
```
