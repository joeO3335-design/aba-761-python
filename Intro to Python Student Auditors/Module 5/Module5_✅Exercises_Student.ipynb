# Module 5- Data Cleaning and Preparation

1. This exercise is a simulation of messy data. The code below generates fake ABA session data with 10,000 samples. 
1. The data is fake, but the concepts and operations we will cover in this module are directly applicable to your work with real data. 
1. If you ever need to make a new dataframe with the fake data, please use this in any cell: `df = pd.DataFrame(data)`

```python
import pandas as pd
import numpy as np
import random
import uuid

# Set the seed for reproducibility
np.random.seed(101214)

# Function to introduce messiness into string values
def make_messy(value):
    """Randomly introduce typos, extra spaces, or different casing."""
    if random.random() < 0.2:  # 20% chance of messiness
        if random.random() < 0.3:
            return value.lower()  # Lowercase variant
        elif random.random() < 0.3:
            return value.upper()  # Uppercase variant
        elif random.random() < 0.2:
            return value + "  "  # Extra spaces
        elif random.random() < 0.2:
            return value.replace(" ", "")  # Remove spaces
        elif random.random() < 0.2:  # Introduce missing values
            return np.nan
        else:
            return value[:3] + random.choice("xyz") + value[4:]  # Introduce typo
    return value

def add_outliers(value):
    if random.random() < 0.2:
        if random.random() < 0.1:
            return value + 100
        elif random.random() < 0.1:
            return value - 100
    
    return value

# Generate fake ABA session that loosely resembles real data with introduced noise. 
num_samples = 10_000
data = {
    "session_id": [uuid.uuid4() for _ in range(num_samples)],
    "date": pd.date_range(start="2025-01-01", periods=num_samples, freq="h").strftime("%Y-%m-%d %H:%M"),
    "client_name": [str(np.random.choice(["Bob", "Rohan", "Milo", "Moana", "Maddie"])) for _ in range(num_samples)],
    "therapist": [str(make_messy(np.random.choice(["Emmet", "Blaze", "Uma", "Lily"]))) for _ in range(num_samples)],
    "target_behavior": [str(make_messy(np.random.choice(["Mand for Item", "Eye Contact", "Functional Play", "Following Directions"]))) for _ in range(num_samples)],
    "reinforcement_strategy": [str(make_messy(np.random.choice(["Token System", "Edible", "Verbal Praise", "Break Time"]))) for _ in range(num_samples)],
    "trials": np.where(np.random.rand(num_samples) < 0.05, np.nan, np.random.randint(5, 20, size=num_samples)),  
    "correct_responses": np.vectorize(add_outliers)(np.random.randint(3, 15, size=num_samples)),
    "prompt_level": [str(make_messy(np.random.choice(["Full", "Partial", "Independent"]))) for _ in range(num_samples)],
    "session_duration": np.random.randint(30, 121, size=num_samples),
    "session_type": [str(make_messy(np.random.choice(["In-Person", "Telehealth"]))) for _ in range(num_samples)],
    "engagement_level": [np.random.choice(["High", "Moderate", "Low"]) for _ in range(num_samples)],
    "challenging_behavior_observed": [str(make_messy(np.random.choice(["Yes", "No"], p=[0.3, 0.7]))) for _ in range(num_samples)],
    "therapist_notes": [
        make_messy(np.random.choice(["Client was focused", "Needed multiple prompts", "Great session!", "Some difficulty transitioning"]))
        for _ in range(num_samples)
    ]
}
# Create DataFrame
df = pd.DataFrame(data)
df 
```
