```python
# Compute the cumulative sum of [1, 2, 3, 4, 5]
cumulative_sum = []
running_sum = 0
numbers = [1, 2, 3, 4, 5]

for number in numbers:
    running_sum += number
    cumulative_sum.append(running_sum)
print(cumulative_sum)

# DJC COMMENT: Also,
arr = np.array([1, 2, 3, 4, 5])
arr.cumsum()
```
