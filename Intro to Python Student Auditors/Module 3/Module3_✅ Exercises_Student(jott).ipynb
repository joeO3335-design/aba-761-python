```python
# Create a NumPy array from a list of integers: [1, 2, 3, 4, 5] 
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
print(arr)
print(type(arr))
```

```python
# Print the shape and data type of the array you created.
arr = np.array([1, 2, 3, 4, 5])
print(arr.shape)
print(arr.dtype)
```

```python
# Create a 2D NumPy array called "arr2d" from [[1.5, -0.1, 3], [0, -3, 6.5]] and print it
arr2d = np.array([[1.5, -0.1, 3], [0, -3, 6.5]])
print(arr2d)
```

```python
# Multiply every element in arr2d by 10
arr2d = np.array([[1.5, -0.1, 3], [0, -3, 6.5]])
arr2d * 10
```

```python
# Create a 3x3 array of zeros
zeros = np.zeros((3,3))
print(zeros)
```

```python
# Create a 4x2 array of ones with dtype int
ones = np.ones((4, 2), dtype=int)
print(ones)
```

```python
# Create a NumPy array containing numbers from 0 to 14
arr = np.arange(15)
print(arr)
```

```python
# Retrieve the element at index 3 from an array called "arr" 
# and where "arr" is created using np.arange(10).
arr = np.arange(10)
print(arr[3])
```

```python
# Slice the first five elements of arr
arr = np.arange(10)
print(arr)
print(arr[:5])
```

```python
# Return only the values greater than 5 from arr.
arr = np.arange(10)
print(arr[arr > 5])
```

```python
# Create a 4x4 array of random numbers from a standard normal distribution
arr = np.random.randn(4, 4)
print(arr)
```

```python
# Create an array and divide all elements by 2
arr = np.array([15, 30, 45, 60, 75])
print(arr/ 2)
```

```python
arr = np.arange(15). reshape(3, 5)
print(arr)

print(arr.T)
```

```python
# Find the element-wise maximum between [3, 7, 2] and [5, 2, 9]
arr1 = np.array([3, 7, 2])
arr2 = np.array([5, 2, 9])
print(np.maximum(arr1, arr2))
```

```python
# Create a 4x4 array filled with random values 
# from the standard normal distribution
# Replace all negative values in the created array with 0
arr = np.random.rand(4, 4)
print("Orignal array:")
print(arr)
arr[arr < 0] = 0
print("\nAfter replacing negatives with 0:")
print(arr)
```

```python
# Compute the cumulative sum of [1, 2, 3, 4, 5]
arr = np.array([1, 2, 3, 4, 5])
print(np.cumsum(arr))
```

```python
# Sort a randomly generated 1D array of size 10
arr = np.random.rand(10)
print("Original:")
print(arr)
print("\nSorted array:")
print(np.sort(arr))
```

```python
# Find unique values in [1, 2, 3, 1, 2, 4, 5]
arr = np.array([1, 2, 3, 1, 2, 4, 5])
print(np.unique(arr))
```

```python
# Generate a 5x5 matrix of random integers between 10 and 50
arr = np.random.randint(10, 51,(5, 5))
print(arr)
```

```python
# Count how many positive numbers exist in a random 10-element array
arr = np.random.randn(10)
print(arr)
print("\nCount of positive numbers:", np.sum(arr > 0))
```

```python
# Add a 1D array [1, 2, 3] to each row of a 3x3 matrix of ones
matrix = np.ones((3, 3), dtype = int)
arr = np.array([1, 2, 3])
print("Matrix:")
print(matrix)
print("\n1D array:", arr)
print("\nResult:")
print(matrix + arr)
```

```python
# Swap axes of a 2x3 array
arr = np.array([[1, 2, 3],
               [4, 5, 6]])
print("Original array (2x3):")
print(arr)
print(arr.swapaxes(0, 1))
```

```python
# Select rows at indices [1, 3, 5] from a 6x4 matrix
matrix = np.random.randint(1, 51, (6, 4))
print("\nSelected rows at indices [1, 3, 5]:")
print(matrix[[1, 3, 5]])
```

```python
# Set all elements greater than 5 in a random array to -1
arr = np.random.randint(1, 11,10)
print(arr)
arr[arr > 5] = -1
print("\nAfter setting elements > 5 to -1:")
print(arr)
```

```python
# Compute the square root of all elements in np.arange(10)
arr = np.arange(10)
print("original array:")
print(arr)
print("\nSquare root of each element:")
print(np.sqrt(arr))
```

```python
# Reshape a 1D array of 12 elements into a 3x4 matrix
arr = np.random.randn(12)
print("\nReshaped to 3 x 4 matrix:")
print(arr.reshape(3, 4))
```

```python
# Compute the dot product of two 3x3 matrices filled with ones
A = np.ones((3, 3))
B = np.ones((3, 3))
print(np.dot(A, B))
```

```python
# Check which values in [1, 2, 3, 4, 5] are greater than 3
arr = np.array([1, 2, 3, 4, 5])
print(arr > 3)
```

```python
# Draw 5 samples from a normal distribution with mean=10, std=2
samples = np.random.normal(10, 2, 5)
print(samples)
```

```python
# Sum across axis 1 for a 4x5 random array
arr = np.random.randint(1, 11, (4, 5))
print("\nSum across axis 1 (row-wise):")
print(arr.sum(axis=1))
```
